```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=42)
```

## Decision Trees

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_clf = DecisionTreeClassifier(max_depth=2)
dt_clf.fit(X_train, y_train)

y_pred = dt_clf.predict(X_test)
y_pred_proba = dt_clf.predict_proba(X_test)
```

```python
dt_clf = DecisionTreeClassifier(max_depth=4)
dt_clf.fit(X_train, y_train)

y_pred = dt_clf.predict(X_test)
y_pred_proba = dt_clf.predict_proba(X_test)
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(20, 10))
plot_tree(dt_clf, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
plt.title('Decision Tree Classifier for Iris Dataset')
plt.show()
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_clf_euclidean = KNeighborsClassifier(n_neighbors=3, metric='euclidean')
knn_clf_euclidean.fit(X_train, y_train)

y_pred_euclidean = knn_clf_euclidean.predict(X_test)
y_pred_proba_euclidean = knn_clf_euclidean.predict_proba(X_test)
```

```python
knn_clf_manhattan = KNeighborsClassifier(n_neighbors=3, metric='manhattan')
knn_clf_manhattan.fit(X_train, y_train)

y_pred_manhattan = knn_clf_manhattan.predict(X_test)
y_pred_proba_manhattan = knn_clf_manhattan.predict_proba(X_test)
```

## Artificial Neural Network

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp_clf_default = MLPClassifier(hidden_layer_sizes=(100,), max_iter=300, random_state=42)
mlp_clf_default.fit(X_train, y_train)

y_pred_default = mlp_clf_default.predict(X_test)
y_pred_proba_default = mlp_clf_default.predict_proba(X_test)
```

```python
mlp_clf_custom = MLPClassifier(hidden_layer_sizes=(50, 30), activation='tanh', max_iter=300, random_state=42)
mlp_clf_custom.fit(X_train, y_train)

y_pred_custom = mlp_clf_custom.predict(X_test)
y_pred_proba_custom = mlp_clf_custom.predict_proba(X_test)
```

## SVM

```python
from sklearn.svm import SVC
```

```python
svm_clf_linear = SVC(kernel='linear', probability=True)
svm_clf_linear.fit(X_train, y_train)

y_pred_linear = svm_clf_linear.predict(X_test)
y_pred_proba_linear = svm_clf_linear.predict_proba(X_test)
```

```python
svm_clf_rbf = SVC(kernel='rbf', probability=True)
svm_clf_rbf.fit(X_train, y_train)

y_pred_rbf = svm_clf_rbf.predict(X_test)
y_pred_proba_rbf = svm_clf_rbf.predict_proba(X_test)
```

## Bayesian

## Logistic Regression
